```python
import cv2
import numpy
from matplotlib import pyplot as plt
```

```python
# Read the image
img = cv2.imread('/home/faqih/INF509-PROJEK-PENGCIT/image/tugas02-face_blurring/image-sample2.jpeg', 0)
```

```python
# Show initial image
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Before")
plt.axis('off')
plt.show()
```

```python
# Apply Canny edge detection
edges = cv2.Canny(img, threshold1=150, threshold2=200)  # Adjust the thresholds as needed 
# Gradien < tres1 = not counted as edge, > tres2 = counted as edge, between tres1 and tres2 = depends on connectivity edges > tres2

# Display the original and edge-detected images
plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(122)
plt.imshow(edges, cmap='gray')
plt.title("Edge Detected Image")
plt.axis('off')

plt.show()
```
