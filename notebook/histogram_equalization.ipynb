```python
import cv2
import numpy
from matplotlib import pyplot as plt
```

```python
# membaca gambar dan mengubahnya menjadi grayscale
img = cv2.imread('img_jelek.jpeg', 0)
```

```python
# Membuat histogram dari gambar asli
# Mengkalkulasikan warna dari gambar asli

histr = cv2.calcHist([img], [0], None, [256], [0, 256])
```

```python
# save hasil plot ke dalam bentuk file

result = plt.plot(histr)
plt.savefig('before.png')
```

```python
print_output = cv2.imread('img_jelek.jpeg')  
plt.imshow(cv2.cvtColor(print_output, cv2.COLOR_BGR2RGB))
plt.title("Before")
plt.axis('off')
```

```python
# plt.gca().clear()

# Ekulisasi histogram dari gambar asli
equ = cv2.equalizeHist(img)

# Membuat histogram dari gambar yang sudah di ekualisasi

cv2.imwrite('after.png', equ)

histr = cv2.calcHist([equ], [0], None, [256], [0, 256])

# Hasil kalulasi warna dari gambar yang sudah di ekualisasi

plt.plot(histr)

# save hasil plot ke dalam bentuk file

result = plt.plot(histr)

plt.savefig('after.png')
```

```python
def grayscale_to_rgb(grayscale_img):
    # Konversi citra grayscale ke citra RGB
    rgb_img = cv2.cvtColor(grayscale_img, cv2.COLOR_GRAY2RGB)
    return rgb_img
```

```python
# save gambar yang sudah di ekualisasi ke dalam bentuk file png 

cv2.imwrite('After.jpeg', equ)

# Baca gambar grayscale yang sudah di ekualisasi
equ = cv2.imread('After.jpeg', cv2.IMREAD_GRAYSCALE)

# Ubah gambar ke dalam bentuk mode RGB
equ_rgb = cv2.cvtColor(equ, cv2.COLOR_GRAY2RGB)

# Simpan gambar sebagai file PNG
cv2.imwrite('After.png', equ_rgb)

# Tampilkan gambar yang sudah diubah menjadi RGB
plt.imshow(cv2.cvtColor(equ_rgb, cv2.COLOR_BGR2RGB))
plt.title("After")
plt.axis('off')
plt.show()
```
