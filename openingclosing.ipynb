```python
# OPENING
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Atur sesuai dengan kebutuhan
OPENING_SIZE = 19

img = cv2.imread('opening.png', 1)
kernel = cv2.getStructuringElement(cv2.MORPH_CROSS,(OPENING_SIZE,OPENING_SIZE))
opening_morph = cv2.morphologyEx(img,cv2.MORPH_OPEN,kernel)
figure, axes = plt.subplots(1, 2, figsize=(8, 4))

ax = axes.ravel()
ax[0].imshow(img)
ax[0].set_title("Original")
ax[1].imshow(opening_morph, cmap=plt.cm.gray)
ax[1].set_title("Opening Morph")
figure.tight_layout()

plt.show()
```

```python
# CLOSING
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Atur sesuai dengan kebutuhan
CLOSING_SIZE = 23

img = cv2.imread('closing.png', 1)
kernel = cv2.getStructuringElement(cv2.MORPH_CROSS,(CLOSING_SIZE,CLOSING_SIZE))
opening = cv2.morphologyEx(img,cv2.MORPH_CLOSE,kernel)
figure, axes = plt.subplots(1, 2, figsize=(8, 4))

ax = axes.ravel()
ax[0].imshow(img)
ax[0].set_title("Original")
ax[1].imshow(opening, cmap=plt.cm.gray)
ax[1].set_title("Closing")
figure.tight_layout()

plt.show()
```
